```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import datetime
matplotlib.rcParams.update({'font.size': 20})
```

```python
filename = "efgb_salinity_1col.csv"
df = pd.read_csv(filename, parse_dates=[0], index_col=0)

df
```

```python
ax = df['24 m SBE'][1:].astype(float).plot( figsize=(10,5) )
times = [str(i) for i in range(2008,2017)]
ax.set_xticks( np.arange(0, 3295-365,365) );
ax.set_xticklabels( times, rotation=30 );
ax.set_title( "Salinity from 24m sensors on East Bank", fontsize=24 )
```
