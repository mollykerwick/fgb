```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
matplotlib.rcParams.update({'font.size': 20})
```

```python
filename = "WBsalinity.csv"
df = pd.read_csv(filename, parse_dates=[0,], index_col=0)
df["7/1":"7/31"]
#df.columns
```

```python
df["WFGB 2016"][].plot()
df["Surface TABS Buoy"][:].plot()
plt.legend(["24m", "Surface"])
plt.x_label("")
```

```python
(df["7/1":"7/31"]).mean()[1:-1].plot(figsize=(15,5))
(df["7/1":"7/31"]).max()[1:-1].plot(figsize=(15,5))
(df["7/1":"7/31"]).min()[1:-1].plot(figsize=(15,5))
smean = (df["Surface TABS Buoy"]["7/1":"7/31"]).mean()
plt.axhline(smean,linestyle="--")
plt.xticks(np.arange(10), df.columns[1:-1], rotation=30)
plt.legend(["Mean","Maximum","Minimum","Surface Average"],loc=(0.9,0.2))
plt.title("Salinity at 24m for July")

plt.ylabel("PSU")
```

```python
(df["WFGB 2012"]["7/1":"7/31"]).plot(figsize=(15,5))
(df["WFGB 2016"]["7/1":"7/31"]).plot(figsize=(15,5))
(df["WFGB 2011"]["7/1":"7/31"]).plot(figsize=(15,5))
(df["WFGB 2010"]["7/1":"7/31"]).plot(figsize=(15,5))
plt.legend(["2012","2016","2011","2010"])
```
